# Mean and Variance for True Measures

```python
import numpy as np
from qmcpy import Uniform, DigitalNetB2

# Create a 2D uniform distribution
tm = Uniform(
    DigitalNetB2(2, seed=7),
    lower_bound=[2, 10],
    upper_bound=[6, 20],
)

print("Mean:", tm.mean)
print("Shape:", tm.mean.shape)

print("Covariance:", tm.covariance)
print("Shape:", tm.covariance.shape)
```
