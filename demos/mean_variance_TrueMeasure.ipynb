# Mean and Variance Methods for True Measure

```python
import numpy as np
from qmcpy import Uniform, DigitalNetB2

# Create a 2D uniform distribution
tm = Uniform(
    DigitalNetB2(2, seed=7),
    lower_bound=[2, 10],
    upper_bound=[6, 20],
)

# Theoretical values from new methods
print("Theoretical mean:", tm.mean())
print("Theoretical variance:", tm.var())

n = 2**16
x = tm(n)
print("Empirical mean:", x.mean(axis=0))
print("Empirical variance:", x.var(axis=0))
```
