Now, we want to start using Kronecker sequences. Kronecker sequences are defined as $x_i = \{ (i \boldsymbol{\alpha} + \boldsymbol{\Delta}) \mod \boldsymbol{1}~|~ \boldsymbol{\alpha} \in (0,1)^d, \boldsymbol{\Delta} \in [0, 1)^d, d \in \mathbb{N} \}$, where $d$ is the amount of dimensions. Recall that to generate samples with Lattice, we do this operation.

```python
lattice = qp.Lattice(2) #The 2 indicates that we want 2-dimensional generated samples
L = lattice.gen_samples(512) #The 512 is saying that we want to generate 512 points.
                             #Note, lattice must generate samples in powers of 2.
print(L)

plt.scatter(L[:,0], L[:, 1], color='red', linestyle='dotted')
plt.title('Lattice')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis((0, 1, 0, 1))
plt.grid(True)
plt.show()
```

Programming Kronecker sequences is just like programming Lattice sequence. In the cell below this is how you generate Kronecker sample points.

```python
kronecker = qp.Kronecker(2)
K = kronecker.gen_samples(512)
print(K)
```

For a visual representation of what Kronecker looks like. Below is a scattered plot like we did for Lattice.

```python
plt.scatter(K[:,0], K[:, 1], color='blue', linestyle='dotted')
plt.title('Kronecker')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis((0, 1, 0, 1))
plt.grid(True)
plt.show()
```

Here is more information on how to operate Kronecker sequences. Suppose that our $\boldsymbol{\alpha} = [\sqrt{2} - 1, \sqrt{3} - 1]$ 

```python
#Kronecker(dimension=1, alpha = None, delta = 0, seed_alpha=None, seed_delta = None)
kronecker = qp.Kronecker(alpha = np.array([np.sqrt(2) - 1, np.sqrt(3) - 1]))
K = kronecker.gen_samples(512)
print(K)
```

As we can see, it out puts the alpha based upon our choice for alpha. We can also choose our delta. Suppose that $boldsymbol{\Delta} = [.1, .2]$. Using the same choice for $\boldsymbol{\alpha}$, we get

```python
kronecker = qp.Kronecker(alpha = np.array([np.sqrt(2) - 1, np.sqrt(3) - 1]), delta = np.array([.1, .2]))
K = kronecker.gen_samples(512)
print(K)
```

You were shown how to input specific values for $\boldsymbol{\alpha}$ and $\boldsymbol{\Delta}$. Now, we need to introduce you to random seed in case if there is something in the program you want to error check but have a hard time because of randomization. 

```python
kronecker = qp.Kronecker(2, seed_alpha = 1, seed_delta = 3)
K = kronecker.gen_samples(512)
print(K)
```

You need to specify the amount of dimensions or else it will assume that you want the amount of dimensions set to 1. And you would need to choose seed_alpha and seed_delta seperately or else either 1 would be set at random. Try playing around with the Kronecker sequences for a better understanding.
