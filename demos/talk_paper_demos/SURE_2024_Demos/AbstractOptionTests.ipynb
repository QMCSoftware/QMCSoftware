# Importing Libraries

The following statements import the necessary libraries needed.

```python
import qmcpy as qp
import numpy as np
import matplotlib.pyplot as plt
```

# Lookback Option Class

The following tests the `LookBackOption` class.

```python
dims = 16
abs_tol = 1e-3

aco = qp.LookBackOption(qp.Sobol(dims), volatility=0.2, start_price=100, interest_rate=.05, call_put='call')
# aco = qp.LookBackOption(volatility=0.2, start_price=100, interest_rate=.05, call_put='put') # Put test
kde, a, b, approx_solution, data, fig, axes = qp.CubQMCSobolG(aco, abs_tol).density_estimation(True)
print("Lookback Option value: %.5f (to within %1.e) with %.f paths"% (approx_solution, abs_tol, data.n_total))
plt.show()
```
