```python
mf = cl.mem_flags
ctx = cl.create_some_context()
prg = cl.Program(ctx,kernelsource).build()
queue = cl.CommandQueue(ctx,properties=cl.command_queue_properties.PROFILING_ENABLE)

rng = np.random.default_rng()

g_d = cl.Buffer(ctx, mf.READ_ONLY | mf.COPY_HOST_PTR, hostbuf=g)
x_d = cl.Buffer(ctx, mf.READ_WRITE, np.dtype(np.float64).itemsize*n*d)
xr_d = cl.Buffer(ctx, mf.WRITE_ONLY, np.dtype(np.float64).itemsize*r*n*d)
shifts_d = cl.Buffer(ctx, mf.READ_ONLY | mf.COPY_HOST_PTR, hostbuf=rng.random((r,d),dtype=np.float64))

# linear 
#   CL
t_linear_cl = time()
lattice_linear = prg.lattice_linear
event_lattice_linear = lattice_linear(queue,(global_size_n,global_size_d),None,np.uint64(n),np.uint64(d),np.uint64(batch_size_n),np.uint64(batch_size_d),g_d,x_d)
event_lattice_linear.wait() 
t_linear_d_cl = (event_lattice_linear.profile.end-event_lattice_linear.profile.start)*1e-9 
x_linear_cl = np.empty((n,d),dtype=np.float64)
cl.enqueue_copy(queue,x_linear_cl,x_d)
t_linear_cl = time()-t_linear_cl
#   QMCPy
t_linear_np = time()
x_linear_np = (g[None,:]*(np.arange(n,dtype=np.uint64)[:,None]/n))%1
t_linear_np = time()-t_linear_np 
assert np.isclose(x_linear_np-x_linear_cl,0).all()

nstart = np.uint64(0)
gc = np.uint8(False)
if not gc: assert (nstart==0 or (nstart&(nstart-1))==0) and ((nstart+n)&(nstart+n-1))==0, "lattice_b2 with natural ordering requires (nstart is 0 or a power of 2) and (nstart+n is a power of 2)"
t_natural_cl = time()
lattice_b2 = prg.lattice_b2
event_lattice_natural = lattice_b2(queue,(global_size_n,global_size_d),None,np.uint64(n),np.uint64(d),np.uint64(batch_size_n),np.uint64(batch_size_d),nstart,gc,g_d,x_d)
event_lattice_natural.wait() 
t_natural_d_cl = (event_lattice_natural.profile.end-event_lattice_natural.profile.start)*1e-9 
x_natural_cl = np.empty((n,d),dtype=np.float64)
cl.enqueue_copy(queue,x_natural_cl,x_d)
t_natural_cl = time()-t_natural_cl 
print("\n%s"%x_natural_cl)

t_rshift_cl = time()
lattice_rshift = prg.lattice_rshift
event_lattice_rshift = lattice_rshift(queue,(global_size_r,global_size_n,global_size_d),None,np.uint64(r),np.uint64(n),np.uint64(d),np.uint64(batch_size_r),np.uint64(batch_size_n),np.uint64(batch_size_d),x_d,shifts_d,xr_d)
event_lattice_rshift.wait() 
t_rshift_d_cl = (event_lattice_rshift.profile.end-event_lattice_rshift.profile.start)*1e-9 
x_rshift_cl = np.empty((r,n,d),dtype=np.float64)
cl.enqueue_copy(queue,x_rshift_cl,xr_d)
t_rshift_cl = time()-t_rshift_cl 
print("\n%s"%x_rshift_cl)

print("\ntiming\n")
print("   t_linear_cl: %.1e\n t_linear_d_cl: %.1e\n   t_linear_np: %.1e\n"%(t_linear_cl,t_linear_d_cl,t_linear_np))
print("  t_natural_cl: %.1e\nt_natural_d_cl: %.1e\n  t_natural_np: NA\n"%(t_natural_cl,t_natural_d_cl))
print("   t_rshift_cl: %.1e\n t_rshift_d_cl: %.1e\n   t_rshift_np: NA\n"%(t_rshift_cl,t_rshift_d_cl))
```
